```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the newly uploaded CSV file into a pandas DataFrame
csv_file_path = 'injury_history(injury_history).csv'
df_injury_history = pd.read_csv(csv_file_path)

# Displaying a summary of the dataset to understand the columns and their data types
df_injury_history.info()

# Displaying the first five rows to get a preview of the dataset
df_injury_history.head()
```

```python
# Convert "Injury Date" to datetime format
df_injury_history['Injury Date'] = pd.to_datetime(df_injury_history['Injury Date'], format='%m/%d/%Y')

# Handling missing values:
# For "Side" and "Severity", we'll fill missing values with 'Unknown' for now.
df_injury_history['Side'].fillna('Unknown', inplace=True)
df_injury_history['Severity'].fillna('Unknown', inplace=True)

# Display the cleaned dataset's summary
df_injury_history.info()

# Display the first five rows to verify the changes
df_injury_history.head()
```

# Injury History Dataset: Exploratory Data Analysis (EDA) Summary

### Data Overview
- **Total Records**: 21
- **Columns**: 10 (Player ID, Name, Group ID, Injury Type, Body Part, Side, Injury Date, Severity, Recovery Time (days), Additional Notes)

### Data Cleaning Steps
1. **Date Conversion**: Converted "Injury Date" from string format to datetime for better analysis.
2. **Missing Values Handling**:
   - "Side" and "Severity" had missing values, which were replaced with "Unknown" to retain data integrity.

### Key Observations and Summary Statistics
- **Injury Types**: Includes various injuries like Muscle Strain, Tendonitis, Strain, and Fracture.
- **Body Parts Affected**: Injuries commonly occurred in parts like Quadriceps, Wrist, Groin, and Shoulder.
- **Recovery Time (days)**: Ranges from a minimum of **11 days** to a maximum of **68 days**.
- **Severity**: Injuries have different severity levels, ranging from Grade 1 to Grade 2, while some are marked as "Unknown".

### Next Steps
1. **Visualizations**:
   - Plot the **frequency of injury types** to identify the most common injuries.
   - Show the **distribution of recovery times** to understand the variance.
   - Analyze **injuries by body part** to see which areas are most affected.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Setting up the plot style for better visual appearance
sns.set(style="whitegrid")

# 1. Frequency of Injury Types
plt.figure(figsize=(10, 6))
sns.countplot(data=df_injury_history, x='Injury Type', order=df_injury_history['Injury Type'].value_counts().index)
plt.title('Frequency of Injury Types')
plt.xlabel('Injury Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()



# 2. Distribution of Recovery Times
plt.figure(figsize=(10, 6))
sns.histplot(df_injury_history['Recovery Time (days)'], kde=True, bins=10)
plt.title('Distribution of Recovery Times (days)')
plt.xlabel('Recovery Time (days)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

# 3. Injuries by Body Part
plt.figure(figsize=(10, 6))
sns.countplot(data=df_injury_history, x='Body Part', order=df_injury_history['Body Part'].value_counts().index)
plt.title('Frequency of Injuries by Body Part')
plt.xlabel('Body Part')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

### Visualizations and Insights
1. **Frequency of Injury Types**:
   - The most common injury types are **Muscle Strain** and **Tendonitis**. These injuries occurred frequently, indicating possible areas for improved training or prevention.

2. **Distribution of Recovery Times**:
   - The recovery time ranges from **11 days** to **68 days**. Most injuries have a recovery period clustered around **20 to 40 days**, as seen from the histogram.

3. **Injuries by Body Part**:
   - The most affected body parts are **Quadriceps**, **Wrist**, and **Groin**. This suggests a need to focus on strengthening or protective measures for these areas to prevent injuries.


```python
# Analyzing the relationship between Injury Type and Severity

# Grouping by 'Injury Type' and 'Severity' to calculate the average recovery time
severity_analysis = df_injury_history.groupby(['Injury Type', 'Severity'])['Recovery Time (days)'].mean().reset_index()

# Visualizing the average recovery time for each injury type by severity level
plt.figure(figsize=(12, 8))
sns.barplot(data=severity_analysis, x='Injury Type', y='Recovery Time (days)', hue='Severity')
plt.title('Average Recovery Time by Injury Type and Severity')
plt.xlabel('Injury Type')
plt.ylabel('Average Recovery Time (days)')
plt.xticks(rotation=45)
plt.legend(title='Severity')
plt.tight_layout()
plt.show()
```

### Analysis of Relationship Between Injury Type and Severity
- The chart shows the **average recovery time** for each injury type categorized by **severity level**.
- **Severity Levels**: The severity is grouped into categories like **Grade 1**, **Grade 2**, and **Unknown**.
- **Recovery Time Patterns**:
  - Injury types with higher severity, such as **Grade 2**, generally have **longer recovery times** compared to less severe categories.
  - For some injury types, such as **Muscle Strain**, recovery time varies significantly depending on the severity level.
- This analysis helps identify which injury types may require more attention and **longer rehabilitation** based on their severity.



```python
# Analyzing the seasonality of injuries by examining the "Injury Date"

# Extracting the month from the "Injury Date" for analysis
df_injury_history['Injury Month'] = df_injury_history['Injury Date'].dt.month

# Grouping by month to count the number of injuries
monthly_injury_counts = df_injury_history['Injury Month'].value_counts().sort_index()

# Visualizing the number of injuries by month
plt.figure(figsize=(10, 6))
sns.barplot(x=monthly_injury_counts.index, y=monthly_injury_counts.values, palette="viridis")
plt.title('Number of Injuries by Month')
plt.xlabel('Month')
plt.ylabel('Number of Injuries')
plt.xticks(ticks=range(12), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.tight_layout()
plt.show()
```

```python
# Extracting the year and month from the "Injury Date" for seasonal analysis
df_injury_history['Injury Year'] = df_injury_history['Injury Date'].dt.year
df_injury_history['Injury Month'] = df_injury_history['Injury Date'].dt.month

# Creating a 'Season' column based on the month
def determine_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

df_injury_history['Season'] = df_injury_history['Injury Month'].apply(determine_season)

# Grouping by 'Season' to count the number of injuries
seasonal_injury_counts = df_injury_history['Season'].value_counts().reindex(['Winter', 'Spring', 'Summer', 'Fall'])

# Visualizing the number of injuries by season
plt.figure(figsize=(10, 6))
sns.barplot(x=seasonal_injury_counts.index, y=seasonal_injury_counts.values, palette="viridis")
plt.title('Number of Injuries by Season')
plt.xlabel('Season')
plt.ylabel('Number of Injuries')
plt.tight_layout()
plt.show()
```

### Seasonality of Injuries
- The chart illustrates the **number of injuries by month** to understand seasonal patterns.
- **Seasonal Trends**:
  - Certain months have higher occurrences of injuries, indicating potential seasonal influences such as weather conditions or increased activity levels.
  - These insights can be useful for **pre-planning injury prevention** measures during high-risk months.

### Injury Trends Over Different Seasons
- The bar chart shows the **number of injuries** that occurred in each **season** (Winter, Spring, Summer, Fall).
- **Seasonal Trends**:
  - The distribution of injuries varies across seasons, with some seasons having higher frequencies.
  - **Spring** and **Fall** tend to have higher injury counts, which may indicate changes in training intensity or activity levels during these times.
  - These insights can help in planning **preventive measures** during high-risk seasons and adjusting training loads accordingly.



```python
# Analyzing the relationship between Injury Type and Group ID to see if there are any patterns in injury occurrence by group

# Grouping by 'Group.Id' and 'Injury Type' to calculate the count of each injury type per group
group_injury_analysis = df_injury_history.groupby(['Group.Id', 'Injury Type']).size().reset_index(name='Count')

# Visualizing the count of each injury type for different groups
plt.figure(figsize=(12, 8))
sns.barplot(data=group_injury_analysis, x='Group.Id', y='Count', hue='Injury Type')
plt.title('Count of Injury Types by Group ID')
plt.xlabel('Group ID')
plt.ylabel('Count of Injuries')
plt.legend(title='Injury Type')
plt.tight_layout()
plt.show()
```

### Analysis of Injury Types by Group ID

- **Group Patterns**:

  - Identifying these patterns can help target specific **injury prevention strategies** tailored to each group.
